```python
from IPython.display import SVG, display
def show_svg(svg):
    display(SVG(svg))
```

Let's create a random permutation over 16 elements, and then convert that into
a routing network:

```python
size = 4
node_width = 12
```

```python
perm = Permutation.random(size)
routed = route_permutation(size, perm)
show_svg(routed.draw_svg(node_width))
```
